```python
import numpy as np
import pandas as pd

data=pd.read_csv(r"C:\Users\MSI GF63\Documents\UiTM Stuff\Semester 4\ISP560 - Machine Learning\Group Project\Income.csv")
print(data)
```

```python
from sklearn.preprocessing import LabelEncoder

data=pd.DataFrame(data,columns=['age','workclass','educational-num','occupation','hours-per-week','income_>50K'])
x=data.iloc[:,0:6].values
y=data.iloc[:,-1].values

le = LabelEncoder()
x[:,1]=le.fit_transform(x[:,1])
x[:,3]=le.fit_transform(x[:,3])
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.6, random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
print(x_train)
print('****************************************************************************')
print(x_test)
```

```python
#Backpropagation process
import tensorflow as tf
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=5, activation = 'relu')) #For 5 input nodes
model.add(tf.keras.layers.Dense(units=10, activation = 'relu')) #For 10 hidden nodes
model.add(tf.keras.layers.Dense(units=2, activation = 'sigmoid')) #For 2 output nodes
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=6, batch_size=128)
```

```python
y_pred = model.predict(x_test)
print(y_pred)

for i in range(len(y_pred)):
    print(np.argmax(y_pred[i]))
```
